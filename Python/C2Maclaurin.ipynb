```python
def maclaurin(func,x,y0=1,maxit=50):
    """
    Display Maclaurin approximations of a function
    input:
        func = terms of the Maclaurin series
        x = value at which the series is evaluated
        y0 = y(0) = the value of y when x=0 (default=1)
        maxit = maximum number of iterations (default=50)
    """
    # the zero-th order is y0
    print(iter:=0, fx:=y0)

    # iterative calculation, from order 1 to maxit
    for iter in range(1,maxit+1):
        fx = fx + func(x,iter)
        print(iter, fx)
```

```python
import math
func = lambda x, iter: x**2*0-1 #x**iter / math.factorial(iter) # exp(x)
maclaurin(func, 0.1, maxit=3) # a func call to estimate e^(0.5)
```

```python
"""
Plot of Maclaurin series of the exponential function
"""
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-3, 5, 100)
f = np.exp(x)
T0 = 1
T1 = T0 + x
T2 = T1 + x**2 / math.factorial(2)
T3 = T2 + x**3 / math.factorial(3)

plt.plot(x,f, '--', label='f')
plt.plot(x,T1, label='T1')
plt.plot(x,T2, label='T2')
plt.plot(x,T3, label='T3')
plt.title('Maclaurin series for exp(x)')
plt.legend(loc='upper left')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.ylim((-5,50))
plt.show()
```
