```python
def euler(x0,xf,y0,dx):
  """
  Input:
    initial condition = (x0,y0), step size = dx
    xf, the value of x at where we want to approximate
  Output:
    values of x and y at each iteration
  """
  i = 0
  x,y,dy = [x0], [y0], [] #dy
  while True:
    x.append(x[i] + dx)
    y.append(y[i] + deriv(x[i], y[i]) * dx)
    dy.append(deriv(x[i], y[i]))
    i = i+1
    if x[i] >= xf-1e-9: break
  return x,y, dy
```

```python
import math
deriv = lambda x,y: (y-x)/(y+x)  # dy/dx = 2y
x,y,dy = euler(0, 2, 1, 0.3333) # x0,xf,y0,dx
print('x=',x)
print('y=',y)
print('dy=',dy)
```
