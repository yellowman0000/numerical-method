```python
import numpy as np
import sympy as sp
```

```python
def bisect(func,xl,xu,es=0.05,maxit=30):
    if func(xl)*func(xu)>0:
        return 'initial estimates do not bracket solution'
    xmold = xl
    for i in range(maxit):
        xm = (xl+xu)/2
        # compute approximate errors
        ea = (xm-xmold)/xm*100
        if abs(ea) < es: break
        # half the search interval
        if func(xm)*func(xl)>0:
            xl = xm
        else:
            xu = xm
        xmold = xm
        print('Iter: ', i+1, 'Xl', xl)
    return xm,func(xm),ea,i+1
```

```python
f = lambda x: 4*(np.sin(2*x))*(1+np.cos(x))

x = sp.symbols('x')
ff = 4 * sp.sin(2*x) * (1 + sp.cos(x))
derf = sp.diff(ff, x)
derff = sp.diff(derf, x)
print(derf)
print(derff)

df = lambda x: 8*np.cos(2*x)*(1+np.cos(x)) - 4*np.sin(x)*np.sin(2*x)
ddf = lambda x: (-16*np.sin(2*x)*(1+np.cos(x))-np.sin(x)*8*np.cos(2*x)) - 4*np.cos(x)*np.sin(2*x) - 8*np.sin(x)*np.cos(2*x)
```

```python
bisect(df, -10, -5)
```

```python
def falsepos(fx, xl, xu, es = 0.05, maxit = 30):
    if(fx(xl)*fx(xu)) > 0:
        return 'Initial not found'
    
    xmold = xl
    for i in range(maxit):
        xm = (fx(xu)*xl - fx(xl)*xu)/(fx(xu)-fx(xl))
        ea = (xm-xmold)/xm*100
        if abs(ea) < es: break
        if(fx(xm)*fx(xl)) > 0:
            xl = xm
        else:
            xu = xm
        xmold = xm
        
        print('Iter: ', i+1, 'Xm: ', xm, 'Ea: ', ea)
    return xm, fx(xm), i+1, ea
```

```python
falsepos(df, -10, -5)
```

```python
def newtraph(fx, df, x0, es = 0.05, maxit = 30):
    for i in range(maxit):
        xr = x0-(fx(x0)/df(x0))
        ea = (xr - x0)/xr*100
        if abs(ea) < es: break
        x0 = xr
        print('Iter: ', i+1, 'xr: ', xr, 'ea: ', ea)
    return xr, fx(xr), ea, i+1 
```
